```
%%writefile /content/Passport-posed-color-from-random-posed-black-n-white-image/temp1.py
import os
import subprocess
from ideepcolor.data import colorize_image as CI
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from shutil import copyfile as cp
import cv2

def run_subp(s):
  process = subprocess.Popen(s.split(), stdout=subprocess.PIPE, stderr=subprocess.PIPE)
  stdout, stderr = process.communicate()

# TODO
def is_grey_scale(img_path):
    return True

def process_image():
  os.chdir("/content/Passport-posed-color-from-random-posed-black-n-white-image/Rotate-and-Render/3ddfa")
  run_subp("python inference.py --img_list example/file_list.txt --img_prefix example/Images --save_dir results")
  # subprocess.run("python inference.py --img_list example/file_list.txt --img_prefix example/Images --save_dir results")
  
  os.chdir("/content/Passport-posed-color-from-random-posed-black-n-white-image/Rotate-and-Render/")
  run_subp("bash experiments/v100_test.sh")
  
  if not is_grey_scale('/content/Passport-posed-color-from-random-posed-black-n-white-image/Rotate-and-Render/results/rs_model/example/aligned/yaw_0.0_test_image.jpg'):
    cp("/content/Passport-posed-color-from-random-posed-black-n-white-image/Rotate-and-Render/results/rs_model/example/aligned/yaw_0.0_test_image.jpg", "/content/Passport-posed-color-from-random-posed-black-n-white-image/static/uploads/result.jpg")
    os.chdir('/content/Passport-posed-color-from-random-posed-black-n-white-image')
    return

  os.chdir("/content/Passport-posed-color-from-random-posed-black-n-white-image/ideepcolor/")
  run_subp("bash ./models/fetch_models.sh")

  gpu_id = 0
  colorModel = CI.ColorizeImageCaffe(Xd=256)
  colorModel.prep_net(gpu_id,'./models/reference_model/deploy_nodist.prototxt','./models/reference_model/model.caffemodel')
  colorModel.load_image('/content/Passport-posed-color-from-random-posed-black-n-white-image/Rotate-and-Render/results/rs_model/example/aligned/yaw_0.0_test_image.jpg') # load an image
  
  mask = np.zeros((1,256,256)) # giving no user points, so mask is all 0's
  input_ab = np.zeros((2,256,256)) # ab values of user points, default to 0 for no input
  img_out = colorModel.net_forward(input_ab,mask) # run model, returns 256x256 image

  img_gray_fullres = colorModel.get_img_gray_fullres() # get grayscale image at fullresolution
  img_out_fullres = colorModel.get_img_fullres() # get image at full resolution

  # show result
  plt.figure(figsize=(7,3))
  plt.imshow(img_out_fullres)
  plt.axis('off')
  plt.savefig("/content/Passport-posed-color-from-random-posed-black-n-white-image/static/uploads/result.jpg")
  os.chdir('/content/Passport-posed-color-from-random-posed-black-n-white-image')
```

```
%%writefile ./app.py
from flask_ngrok import run_with_ngrok
from flask import Flask
from flask import make_response

app = Flask(__name__)
UPLOAD_FOLDER = 'static/uploads'
app.config['UPLOAD_FOLDER'] = UPLOAD_FOLDER
app.secret_key = "secret key"
run_with_ngrok(app)   #starts ngrok when the app is run

import os
import urllib.request
from flask import Flask, flash, request, redirect, url_for, render_template
from werkzeug.utils import secure_filename
from shutil import copyfile as cp
from temp1 import process_image
from flask import session
import os.path, os

ALLOWED_EXTENSIONS = set(['png', 'jpg', 'jpeg'])


def allowed_file(filename):
	return '.' in filename and filename.rsplit('.', 1)[1].lower() in ALLOWED_EXTENSIONS
		
@app.route('/')
def upload_form(): 
	if os.path.isfile('./static/uploads/test_image.jpg'):
		os.remove('./static/uploads/test_image.jpg')
	if os.path.isfile('./static/uploads/result.jpg'):
		os.remove('./static/uploads/result.jpg')
	
	return render_template('upload.html')

@app.route('/upload', methods=['POST'])
def upload_image():
	if 'file' not in request.files:
		flash('No file part')
		return redirect(request.url)
	file = request.files['file']
	if file.filename == '':
		flash('No image selected for uploading')
		return redirect(request.url)
	if file and allowed_file(file.filename):
		filename = secure_filename(file.filename)
		filename = "test_image.jpg"
		file.save(os.path.join(app.config['UPLOAD_FOLDER'], filename))
		# flash('Image successfully uploaded and displayed')
		return render_template('upload.html', filename=filename)
	else:
		flash('Allowed image types are -> png, jpg, jpeg')
		return redirect(request.url)
	

@app.route('/display/<filename>')
def display_image(filename):
	return redirect(url_for('static', filename='uploads/' + filename), code=301)

@app.route('/custom', methods=['GET'])
def custom():
	cp("/content/Passport-posed-color-from-random-posed-black-n-white-image/static/uploads/test_image.jpg", "/content/Passport-posed-color-from-random-posed-black-n-white-image/Rotate-and-Render/3ddfa/example/Images/test_image.jpg")
	process_image()
	filename = "result.jpg"
	return redirect(url_for('static', filename='uploads/' + filename), code=301)
 
app.run()
```
